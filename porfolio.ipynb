```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("C:/Users/Hp/Downloads/archive/googleplaystore.csv")
df1.shape
```

```python
#Dropping duplicated rows
#On play store, two apps may have same name, but all the size, installs, rating, reviews, price need not be same. so using these categories, we will drop the duplicates in the data
df1.drop_duplicates(['App','Size','Installs','Reviews','Rating','Price','Android Ver'], inplace=True)

#Dropping null rows
df1=df1.dropna()

df1.head()
```

```python
#Plotting a graph to show the amount of apps per category

plt.title("Amount of apps per category")
plt.xlabel("Category")
plt.ylabel("Count")
valuecounts=df1['Category'].value_counts()

valuecounts.plot.bar()
```

```python
#Plotting a graph to show the Frequency distribution per Rating

plt.grid()
plt.title("Frequency per Rating" )
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.hist(df1['Rating'])
```

```python
#Plotting a pi chart to show the percentage of apps that are free vs the ones that are paid

valuecounts=df1['Type'].value_counts()
explode = (0.05, 0.05) 
valuecounts.plot.pie(autopct='%1.1f%%',explode=explode)
plt.title('Percentage distribution of apps that are free vs the ones that are paid')
```

```python
#Plotting a bar graph to show the amount of apps per each Rating

plt.title("Amount of apps per each Rating")
plt.xlabel("count")
plt.ylabel("Content Rating")
valuecounts=df1['Content Rating'].value_counts()
plt.xscale('log')
valuecounts.plot.barh()
```

```python
#Plotting a bar graph to show the Number of Installations per each grading

plt.title("Number of Installations")
plt.xlabel("Classes of Installations")
plt.ylabel("Numbr of Installs")
valuecounts=df1['Installs'].value_counts()

valuecounts.plot.bar()
```
